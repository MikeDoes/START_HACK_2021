## Functions use

```python
def conver_df(filename):
    df = []
    with h5py.File(filename, "r") as f:
        columns = list(f.keys())
        for c in tqdm(columns):
            df.append(list(f[c]))

    df = pd.DataFrame(df).T
    df.columns = columns
    return df

def clean_df(df):
    data = []
    for j, (r,l) in enumerate(tqdm(list(zip(list(dict(df['acc_RH']).values()), list(dict(df['acc_LH']).values()))))):
        if r is not None:
            data.append([r[0], r[1], r[2], r[3], l[1], l[2], l[3]])
    data = pd.DataFrame(data)
    data.columns = ['Time Step', 'Right AP', 'Right UR', 'Right DP', 'Left AP', 'Left UR', 'Left DP']
    return data
```

### Load json

```python
# JSON file 
f = open ('datasets/traverse.json', "r") 
  
# Reading from file 
data = json.loads(f.read()) 
```

```python
data
```

```python
position = []
for d in data:
    aux = d['keypoints']
    p = [None, None, None, None, None, None]
    for a in aux:
        if a['part'] == 'leftShoulder':
            p[0] = [a['position']['x'], a['position']['y']]
        elif a['part'] == 'rightShoulder':
            p[1] = [a['position']['x'], a['position']['y']]
        elif a['part'] == 'leftWrist':
            p[2] = [a['position']['x'], a['position']['y']]
        elif a['part'] == 'rightWrist':
            p[3] = [a['position']['x'], a['position']['y']]
        elif a['part'] == 'leftElbow':
            p[4] = [a['position']['x'], a['position']['y']]
        elif a['part'] == 'rightElbow':
            p[5] = [a['position']['x'], a['position']['y']]
    position.append(p.copy())
```

```python
df = pd.DataFrame(position, columns=['leftShoulder', 'rightShoulder', 'leftWrist', 
                                     'rightWrist', 'leftElbow', 'rightElbow'])
df
```

```python
df1 = pd.concat([pd.DataFrame(df[c].values.tolist()) for c in df.columns], 
                 axis=1, 
                 keys=df.columns)
df1.columns = ['{}{}'.format(i, j) for i, j in df1.columns]

col = df1.loc[: , ["leftShoulder0","rightShoulder0"]]
df1['shoulder0'] = col.mean(axis=1)
col = df1.loc[: , ["leftShoulder1","rightShoulder1"]]
df1['shoulder1'] = col.mean(axis=1)

left_wrist_norm0 = df1.loc[:, 'leftWrist0'] - df1.loc[: , 'shoulder0']
right_wrist_norm0 = df1.loc[:, 'rightWrist0'] - df1.loc[: , 'shoulder0']
left_wrist_norm1 = df1.loc[:, 'leftWrist1'] - df1.loc[: , 'shoulder1']
right_wrist_norm1 = df1.loc[:, 'rightWrist1'] - df1.loc[: , 'shoulder1']

left_elbow_norm0 = df1.loc[:, 'leftElbow0'] - df1.loc[: , 'shoulder0']
right_elbow_norm0 = df1.loc[:, 'rightElbow0'] - df1.loc[: , 'shoulder0']
left_elbow_norm1 = df1.loc[:, 'leftElbow1'] - df1.loc[: , 'shoulder1']
right_elbow_norm1 = df1.loc[:, 'rightElbow1'] - df1.loc[: , 'shoulder1']

left_shoulder_norm0 = df1.loc[:, 'leftShoulder0'] - df1.loc[: , 'shoulder0']
right_shoulder_norm0 = df1.loc[:, 'rightShoulder0'] - df1.loc[: , 'shoulder0']
left_shoulder_norm1 = df1.loc[:, 'leftShoulder1'] - df1.loc[: , 'shoulder1']
right_shoulder_norm1 = df1.loc[:, 'rightShoulder1'] - df1.loc[: , 'shoulder1']

df1['lwn0'] = left_wrist_norm0
df1['rwn0'] = right_wrist_norm0
df1['lwn1'] = left_wrist_norm1
df1['rwn1'] = right_wrist_norm1

df1['len0'] = left_elbow_norm0
df1['ren0'] = right_elbow_norm0
df1['len1'] = left_elbow_norm1
df1['ren1'] = right_elbow_norm1

df1['lsn0'] = left_shoulder_norm0
df1['rsn0'] = right_shoulder_norm0
df1['lsn1'] = left_shoulder_norm1
df1['rsn1'] = right_shoulder_norm1

df1[['lwn0', 'lwn1']]
```

```python
def draw_arms(coordinates):
    x_l_forearm = [coordinates.iloc[0], coordinates.iloc[4]]
    y_l_forearm = [coordinates.iloc[2], coordinates.iloc[6]]
    x_l_arm = [coordinates.iloc[4], coordinates.iloc[8]]
    y_l_arm = [coordinates.iloc[6], coordinates.iloc[10]]
    plt.plot(x_l_forearm, y_l_forearm, color = 'red')
    plt.plot(x_l_arm, y_l_arm, color = 'red')
    
    x_r_forearm = [coordinates.iloc[1], coordinates.iloc[5]]
    y_r_forearm = [coordinates.iloc[3], coordinates.iloc[7]]
    x_r_arm = [coordinates.iloc[5], coordinates.iloc[9]]
    y_r_arm = [coordinates.iloc[7], coordinates.iloc[11]]
    plt.plot(x_r_forearm, y_r_forearm, color = 'blue')
    plt.plot(x_r_arm, y_r_arm, color = 'blue')
    #plt.show()
```

## Clusterize the poses of the user

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
x = StandardScaler().fit_transform(features)
```

```python
pca = PCA(n_components=2)
res = pca.fit(x)
```

```python
X = pca.transform(x)
```

```python
X = x
```

```python
n_clus = 4
```

```python
kmeans = KMeans(n_clusters = n_clus, random_state=0).fit(X)
```

```python
plt.scatter(features.iloc[:,0], features.iloc[:,2], c = kmeans.labels_)
```
